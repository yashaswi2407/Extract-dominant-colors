# Dominant Color Extraction for image segmentation 

```python
import numpy as np
import pandas as pd
import cv2 
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("rome.jpg") #it read in BGR format
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
org_shape=img.shape
print(org_shape)
```

```python
plt.imshow(img)
```

```python
#flatten each channel of the image
all_pixel=img.reshape((-1,3))
print(all_pixel.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
dominant_colors=4
km=KMeans(n_clusters=dominant_colors)
km.fit(all_pixel)
```

```python
centers=km.cluster_centers_
```

```python
center=np.array(centers,dtype='uint8')
```

```python
center
#this it the value of the color we got from the kmeans algo
```

```python
#plot what all this color are these?
i=1
plt.figure(0,figsize=(4,2))

colors=[]
for each_col in centers:
    plt.subplot(1,4,i)
    i+=1
    colors.append(each_col)
    #color swatch
    a=np.zeros((100,100,3),dtype='uint8')
    a[:,:,:]=each_col
    plt.imshow(a)
plt.show()
```

# Segmenting our original image

```python
new_img=np.zeros((183*275,3),dtype='uint8')
print(new_img.shape)
```

```python
colors #0,1,2,3 respectively
```

```python
km.labels_
```

```python
for ix in range(new_img.shape[0]):
    new_img[ix]=colors[km.labels_[ix]]
new_img=new_img.reshape((org_shape))
plt.imshow(new_img)
plt.show()
```
